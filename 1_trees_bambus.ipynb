```python
tree = toytree.tree("./analysis-raxml/RAxML_bipartitions.bambu1")

#import real names as dict
names = pandas.read_csv("nombres_especies_RAD.csv", index_col=0, squeeze=True, usecols=[0,1])
names = names.to_dict() # put them in dict form

# new_tips = [f"{names[tip]}_{tip[3:]}" for tip in tree.get_tip_labels()]

# set dict with new tips (real name and code)
new_tips = {tip:f"{names[tip]}_{tip[3:]}" for tip in tree.get_tip_labels()}

# alter the tree itself to mantain new names
tree = tree.set_node_values(feature="name", values=new_tips)

# root tree in Guadua
tree = tree.root(wildcard="G.")

# collapse low support
tree = tree.collapse_nodes(min_support=50)

# set colors for very bad samples (few reads) AFTER ROOTING
new_colors = ["red" if tip in [f"{names[i]}_{i[3:]}" for i in bad_samples] else "black" for tip in tree.get_tip_labels()]

c,a,m = tree.draw(width=1000, 
                  node_labels=tree.get_node_values("support"), 
                  node_sizes=15,
                  tip_labels_colors=new_colors,
                 )
```

```python
import toyplot.svg
toyplot.svg.render(c, f"bambu1.svg")
```

```python
import toyplot.svg
toyplot.svg.render(c, f"bambus7_downMissingdata.svg")
```

```python
import toytree
```

```python
tre = toytree.rtree.baltree(4)
```

```python
colordict = {'r1': 'red',
 'r2': 'blue',
 'r3': 'green',
 'r0': 'pink'}
```

```python
# you need a colordict, where the key is the name of the sample and the item is the color, for example: 'microcarpum': '#FFC000',

color_labels = []

#Populate color_labels based on the colordict
for i in tre.get_tip_labels():  
    result = "Black" #set black as default
    for key, item in colordict.items():
        if i.find(key) > -1:
            result = item
    color_labels.append(result)
    
#after this loop you can be sure that your color_labels will have the same order in your tree
color_labels
```

```python
tre.draw(tip_labels_colors=color_labels)
```
